```python
# Import Libraries and Dependencies
import pandas as pd
```

### 1. Combine and Clean the Data
#### Import CSVs

```python
# Read the CSV files into DataFrames.
athletic_sales_2020_df = pd.read_csv('Resources/athletic_sales_2020.csv')
athletic_sales_2021_df = pd.read_csv('Resources/athletic_sales_2021.csv')
```

```python
# Display the 2020 sales DataFrame
athletic_sales_2020_df.head()
```

```python
# Display the 2021 sales DataFrame
athletic_sales_2021_df.head()
```

#### Check the data types of each DataFrame

```python
# Check the 2020 sales data types.
athletic_sales_2020_df.info()
```

```python
# Check the 2021 sales data types.
athletic_sales_2021_df.info()
```

#### Combine the sales data by rows.

```python
# Combine the 2020 and 2021 sales DataFrames on the rows and reset the index.
joined_data_rows_df = pd.concat([athletic_sales_2020_df, athletic_sales_2021_df], axis="rows", join="inner")
joined_data_rows_df
```

```python
# Check if any values are null.
null_counts = joined_data_rows_df.isnull().sum()
null_counts
```

```python
# Check the data type of each column
joined_data_rows_df.info()
```

```python
# Convert the "invoice_date" to a datetime datatype
joined_data_rows_df['invoice_date'] = pd.to_datetime(joined_data_rows_df['invoice_date'])
joined_data_rows_df
```

```python
# Confirm that the "invoice_date" data type has been changed.
joined_data_rows_df.info()
```

### 2. Determine which Region Sold the Most Products

#### Using `groupby`

```python
# Show the number products sold for region, state, and city.
num_products_df = joined_data_rows_df.groupby(["region","state","city"])[["units_sold"]].sum()

# Rename the sum to "Total_Products_Sold".
num_products_df = num_products_df.rename(columns={'units_sold': 'Total_Products_Sold'})

# Show the top 5 results.
sorted_num_products_df = num_products_df.sort_values(by='Total_Products_Sold', ascending=False)
sorted_num_products_df.head()
```

#### Using `pivot_table`

```python
# Show the number products sold for region, state, and city.
num_products_pivot_df = pd.pivot_table(joined_data_rows_df, 
                                       values='units_sold',
                                       index=['region', 'state', 'city'], 
                                       aggfunc='sum')

# Rename the "units_sold" column to "Total_Products_Sold"
num_products_pivot_df = num_products_pivot_df.rename(columns={'units_sold': 'Total_Products_Sold'})

# Show the top 5 results.
sorted_num_products_pivot_df = num_products_pivot_df.sort_values(by='Total_Products_Sold', ascending=False)
sorted_num_products_pivot_df.head()
```

### 3. Determine which Region had the Most Sales

#### Using `groupby`

```python
# Show the total sales for the products sold for each region, state, and city.
total_sales_df = joined_data_rows_df.groupby(["region","state","city"])[["total_sales"]].sum()

# Rename the "total_sales" column to "Total Sales"
total_sales_df = total_sales_df.rename(columns={'total_sales': 'Total Sales'})

# Show the top 5 results.
sorted_total_sales_df = total_sales_df.sort_values(by='Total Sales', ascending=False)
sorted_total_sales_df.head()
```

#### Using `pivot_table`

```python
# Show the total sales for the products sold for each region, state, and city.
total_sales_pivot_df = pd.pivot_table(joined_data_rows_df, 
                                       values='total_sales',
                                       index=['region', 'state', 'city'], 
                                       aggfunc='sum')

# Optional: Rename the "total_sales" column to "Total Sales"
total_sales_pivot_df = total_sales_pivot_df.rename(columns={'total_sales': 'Total Sales'})

# Show the top 5 results.
sorted_total_sales_pivot = total_sales_pivot_df.sort_values(by='Total Sales', ascending=False)
sorted_total_sales_pivot.head()
```

### 4. Determine which Retailer had the Most Sales

#### Using `groupby`

```python
# Show the total sales for the products sold for each retailer, region, state, and city.
total_sales_retailer_df = joined_data_rows_df.groupby(["retailer","region","state","city"])[["total_sales"]].sum()

# Rename the "total_sales" column to "Total Sales"
total_sales_retailer_df = total_sales_retailer_df.rename(columns={'total_sales': 'Total Sales'})

# Show the top 5 results.
total_sales_retailer_df.head()
```

#### Using `pivot_table`

```python
# Show the total sales for the products sold for each retailer, region, state, and city.
total_sales_retailer_pivot_df = pd.pivot_table(joined_data_rows_df, 
                                       values='total_sales',
                                       index=['retailer','region', 'state', 'city'], 
                                       aggfunc='sum')

# Optional: Rename the "total_sales" column to "Total Sales"
total_sales_retailer_pivot_df = total_sales_retailer_pivot_df.rename(columns={'total_sales': 'Total Sales'})

# Show the top 5 results.
sorted_total_sales_retailer_pivot = total_sales_retailer_pivot_df.sort_values(by='Total Sales', ascending=False)
sorted_total_sales_retailer_pivot
```

### 5. Determine which Retailer Sold the Most Women's Athletic Footwear

```python
# Filter the sales data to get the women's athletic footwear sales data.
# Step 1: find the product type name to sort by
print(joined_data_rows_df['product'].unique())

# Step 2
filter_womens_athletic_footwear_df = joined_data_rows_df.loc[joined_data_rows_df["product"] == "Women's Athletic Footwear", :]
filter_womens_athletic_footwear_df.head()
```

#### Using `groupby`

```python
# Show the total number of women's athletic footwear sold for each retailer, region, state, and city.
womens_athletic_footwear_units_sold_df = filter_womens_athletic_footwear_df.groupby(["retailer","region","state","city"])[["units_sold"]].sum()

# Rename the "units_sold" column to "Womens_Footwear_Units_Sold"
womens_athletic_footwear_units_sold_df = womens_athletic_footwear_units_sold_df.rename(columns={'units_sold': 'Womens_Footwear_Units_Sold'})

# Show the top 5 results.
womens_athletic_footwear_units_sold_df = womens_athletic_footwear_units_sold_df.sort_values(by='Womens_Footwear_Units_Sold', ascending=False)
womens_athletic_footwear_units_sold_df.head()
```

#### Using `pivot_table`

```python
# Show the total number of women's athletic footwear sold for each retailer, region, state, and city.
womens_athletic_footwear_units_sold_pivot_df = pd.pivot_table(filter_womens_athletic_footwear_df, 
                                       values='units_sold',
                                       index=['retailer','region', 'state', 'city'], 
                                       aggfunc='sum')

# Rename the "units_sold" column to "Womens_Footwear_Units_Sold"
womens_athletic_footwear_units_sold_pivot_df = womens_athletic_footwear_units_sold_pivot_df.rename(columns={'units_sold': 'Womens_Footwear_Units_Sold'})

# Show the top 5 results.
womens_athletic_footwear_units_sold_pivot_df = womens_athletic_footwear_units_sold_pivot_df.sort_values(by='Womens_Footwear_Units_Sold', ascending=False)
womens_athletic_footwear_units_sold_pivot_df.head()
```
